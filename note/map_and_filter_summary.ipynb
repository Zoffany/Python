# `map` と `filter` の直感的なまとめ

このノートブックでは、Pythonの重要な組み込み関数である **`map`** と **`filter`** について、直感的に理解できるように整理しています。

## 💡 全体像のイメージ

両方とも「リストなどの要素を1個ずつ処理する」ためのショートカット機能（イテレータ）ですが、役割が異なります。

| 関数 | イメージ | 役割 | 出力データの個数 |
| :--- | :--- | :--- | :--- |
| **`map`** | **変換マシン** | 全要素に同じ加工ルールを適用する | 元のリストと**同じ個数** |
| **`filter`** | **選別ふるい** | 条件に合う（Trueになる）要素だけを残す | 元のリストより**減る（または同じ）** |

---
## 1. `map` （変換マシン）

**「全員に、この処理（関数）を一括で適用して！」** という時に使います。

### 🛠️ 動く例：リストの数字をすべて2倍にする
※ 下のセルを実行して、結果を確認してみましょう！

```python
numbers = [1, 2, 3, 4, 5]

# 各要素を2倍にするラムダを適用し、list()で実体化
doubled = list(map(lambda x: x * 2, numbers))
print(doubled)
```

### 🛠️ 動く例：既存の関数（str）を使って文字列にする
すでに用意されている関数（strなど）を適用する場合は、ラムダすら不要になります。

```python
numbers = [1, 2, 3]

# 全体を文字列にする
stringified = list(map(str, numbers))
print(stringified)
```

---
## 2. `filter` （選別ふるい）

**「この条件（TrueかFalseを返す関数）に合うものだけを残して！」** という時に使います。
条件式が `True` になる要素だけがフィルターを通り抜け、`False` のものは脱落します。

### 🛠️ 動く例：偶数（2で割り切れる数）だけを取り出す

```python
numbers = [1, 2, 3, 4, 5, 6]

# 「2で割った余りが0か？」という条件がTrueの要素だけを残す
evens = list(filter(lambda x: x % 2 == 0, numbers))
print(evens)
```

### 🛠️ 実践例：異常検知（しきい値を超える温度データだけを抽出する）
製造業のセンサーデータ分析で、温度が100度を超えている異常データだけを取り出したい場合のイメージです。

```python
temperatures = [98.5, 102.1, 99.0, 105.3, 97.8]

# 100度より高いデータ（異常値）だけを抽出
anomalies = list(filter(lambda temp: temp > 100.0, temperatures))
print(anomalies)
```

---
## 3. ⚠️ 超重要：`map` と `filter` の決定的な違い

同じ条件式（例：`lambda x: x > 2`）を `map` と `filter` に渡したとき、返ってくる結果がまったく異なります。
ここが一番間違えやすいポイントです！

### 🛠️ 実際に比較してみましょう

```python
data = [1, 2, 3, 4]

# 1. mapに渡した場合（「x > 2 か？」という判定結果に全員を「変換」する）
map_result = list(map(lambda x: x > 2, data))
print("mapの結果:   ", map_result)  # True/False のリストになる

# 2. filterに渡した場合（「x > 2 か？」が True のものだけを「残す」）
filter_result = list(filter(lambda x: x > 2, data))
print("filterの結果:", filter_result)  # 基準を満たした数値だけのリストになる
```

## ✍️ まとめ

1. **`map` は「変換マシン」**: 全員の形を変える（例: 数値 ➡ 文字列、数値 ➡ True/False）
2. **`filter` は「選別ふるい」**: 条件に合う実体だけを残して削る（例: 異常値だけ、偶数だけ）
3. **どちらも「保留状態（イテレータ）」で返る**ため、中身をリストで確認したい時は `list(...)` で囲む。

これで `map` と `filter` の基本と違いはバッチリです！
