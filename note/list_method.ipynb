```python
fruits=['apple','banana']
rts=['wc','sc','aoe']
```

---
# ***.append()***
末尾に一つ追加する。リスト渡すとネスト化

```python
fruits.append('lemon')
print(fruits)
```

```python
fruits.append(['X','Y'])
print(fruits)
```

---
# ***lsA.extend(lsB)***

末尾に別リスト結合。リスト渡すけどフラット

```python
fruits.extend(rts)
print(fruits)
```

***.append()   .extend()***

違い

```python
A=['1','2','3']
B=['4','5','6']

A.append(B)
print(A)

A=['1','2','3']
A.extend(B)
print(A)
```

---
# ***.insert(index,value)***

index位置に入れる。元のvalueは後ろにずれる

押し出す処理が全てに走るので重い

```python
fruits.insert(2,('Kudamono'))
print(fruits)
```

---
# ***.remove(value)***

一度に最初にヒットするひとつだけ消去

直接指定で手間だからこれよりPopかPandasおすすめ

```python
fruits.remove('wc')
print(fruits)
```

```python
fruits.remove(fruits[4])
print(fruits)
```

---
# ***.pop()***

index指定して消去。しないと末尾消去
消去したvalueも取れる

```python
x=fruits.pop(2)
print(fruits)
print(x)
```

```python
fruits.pop()
print(fruits)
```

---
# ***sorted()***

並び替え。非破壊的、こっち使いたいね

メソッドじゃないよ。引数で降順。Trueの頭大文字！

```python
ls=[4,3,2,1]
ls1=sorted(ls)
print(ls1)
```

# ***.sort()***

並び替え。引数で降順

```python
ls=[2,4,1,3]

ls.sort()
print(ls)

ls.sort(reverse=True)
print(ls)
```

---
# ***.reverse()***

ただひっくり返すだけ

```python
ls.reverse()
print(ls)

ls2=[4,33,2,3]
ls2.reverse()
print(ls2)
```

---
# ***.index()***

value渡してindex返す。存在しない引数はエラー

```python
print(fruits)
fruits.index('sc')
```

```python
#Getみたいな安全版がないから実務ではIf使うよ

if 'apple' in fruits:
  print(fruits.index('sc'))
```

# ***.count()***
value渡してその個数を返す

```python
fruits.count('sc')
```

# ***.copy()***

コピーします

```python
print(rts)
strat=rts.copy()
game=rts
rts.append('x')  #game.append('x')と同義

#コピーは影響なし
print(strat)
print(game)
```
