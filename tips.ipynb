```python
def sum_strings(list1):
  ls=''
  for x in range(len(list1)):
    strr=str(list1[x])  #forループ、ifの前に条件挟めるよ
    if 0<=x<(len(list1))-2:
      ls=ls+strr+', '  #文字列で出力したいときは’’に＋
    elif x==(len(list1))-2:
      ls=ls+strr+' and '  #リストで出したいならアペンドかインデックス指定して代入
    else:
       ls=ls+strr
  return ls
```

```python
user_info = {'name': 'Taro', 'age': 25}

for key in user_info:
    # 1. まず key に 'name' が入る
    # 2. 次に key に 'age' が入る

    # 中身（値）を取り出すときは、こう書く
    value = user_info[key]
    print(f"{key} は {value} です")
```

***リストも辞書も取りだしたらただの値***
